```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
scaled_data = scaler.fit_transform(df[numeric_cols])
```

```python
plt.figure(figsize=(16,12))
for i, col in enumerate(numeric_cols[:7],1):
    plt.subplot(4,2,i)
    sns.histplot(df[col], kde=True, bins=30, color='blue')
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16,12))
for i, col in enumerate(numeric_cols[:7],1):
    plt.subplot(4,2,i)
    sns.boxplot(x=df[col], color='green')
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_data)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
linked = linkage(scaled_data, method='ward')
```

```python
plt.figure(figsize=(12,8))
dendrogram(linked)
plt.title('Hierarchical Clustering Dendogram')
plt.show()
```

```python
hierarchical_labels = fcluster(linked, t=4, criterion='maxclust')
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=2, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled_data)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(pca_data[:,0], pca_data[:,1], c=kmeans_labels, cmap='viridis', label='K-Means')
plt.title('K-Means Cluster (PCA Reduced Data)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Compoennt 2')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(pca_data[:,0], pca_data[:,1], c=hierarchical_labels, cmap='plasma', label='Hierarchical')
plt.title('Hierarcical Cluster (PCA Reduced Data)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Compoennt 2')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(pca_data[:,0], pca_data[:,1], c=dbscan_labels, cmap='coolwarm', label='DBSCAN')
plt.title('DBSCAN Cluster (PCA Reduced Data)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Compoennt 2')
plt.colorbar()
plt.show()
```

```python
print('K-Means Clustering Labels: ', np.unique(kmeans_labels))
print('Hierarchical Clustering Labels: ', np.unique(hierarchical_labels))
print('DBSCAN Clustering Labels: ', np.unique(dbscan_labels))
```
